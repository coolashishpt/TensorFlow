```python
from random import sample
def train_test_creation(x, data, toPred): 
    indices = sample(range(data.shape[0]),int(x * data.shape[0])) 
    indices = np.sort(indices, axis=None) 
  
    index = np.arange(data.shape[0]) 
    reverse_index = np.delete(index, indices,0)
  
    train_toUse = data[indices]
    train_toPred = toPred[indices]
    test_toUse = data[reverse_index]
    test_toPred = toPred[reverse_index]

    return train_toUse, train_toPred, test_toUse, test_toPred

xtrain, ytrain, xtest, ytest = train_test_creation(0.7, feature, labels)
print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape )
```

```python
def initialize_parameters():
    tf.set_random_seed(1)                             
    
    # shape = (4, 4, 3, 16) which mean is 4X4 filter with depth 3 and number of filter is 16
    W1 = tf.Variable(tf.truncated_normal([4, 4, 3, 16], stddev=0.1))
    W2 = tf.Variable(tf.truncated_normal([2, 2, 16, 32], stddev=0.1))
    W3 = tf.Variable(tf.truncated_normal([2, 2, 32, 64], stddev=0.1))
    lW1 = tf.Variable(tf.truncated_normal([15 * 15 * 64, 1024], stddev=0.1))
    lW2 = tf.Variable(tf.truncated_normal([1024, 1024], stddev=0.1))
    WOUT = tf.Variable(tf.truncated_normal([1024, 8], stddev=0.1))
    
    
    B1 = tf.Variable(tf.constant(0.1, shape=[16]))
    B2 = tf.Variable(tf.constant(0.1, shape=[32]))
    B3 = tf.Variable(tf.constant(0.1, shape=[64]))
    lB1 = tf.Variable(tf.constant(0.1, shape=[1024]))
    lB2 = tf.Variable(tf.constant(0.1, shape=[1024]))
    BOUT = tf.Variable(tf.constant(0.1, shape=[8]))
    
    
    parameters = {
        "w1": W1,
        "w2": W2,
        "w3": W3,
        "lw1": lW1,
        "lw2": lW2,
        
        "b1": B1,
        "b2": B2,
        "b3": B3,
        "lb1": lB1,
        "lb2": lB2,
        
        "wout": WOUT,
        "bout": BOUT
    }
    return parameters
```

```python
def forward_prop(X, parameters):
    
    
    w1 = parameters["w1"]
    w2 = parameters["w2"]
    w3 = parameters["w3"]
    b1 = parameters["b1"]
    b2 = parameters["b2"]
    b3 = parameters["b3"]
    
    lw1 = parameters["lw1"]
    lw2 = parameters["lw2"]
    lb1 = parameters["lb1"]
    lb2 = parameters["lb2"]
    
    wout = parameters["wout"]
    bout = parameters["bout"]
    
    
    
    #LAYER 1
    conv1 = tf.nn.conv2d(X, w1, strides=[1,1,1,1], padding="SAME") + b1
    # Relu
    a1 = tf.nn.relu(conv1)
    # max pooling
    p1 = tf.nn.max_pool(a1, ksize=[1,2,2,1], strides=[1,2,2,1], padding="SAME")
    
    
    #LAYER 2
    conv2 = tf.nn.conv2d(p1, w2, strides=[1,1,1,1], padding="SAME") + b2
    # Relu
    a2 = tf.nn.relu(conv2)
    # max pooling
    p2 = tf.nn.max_pool(a2, ksize=[1,2,2,1], strides=[1,2,2,1], padding="SAME")
    
    
    #LAYER 3
    conv3 = tf.nn.conv2d(p2, w3, strides=[1,1,1,1], padding="SAME") + b3
    # Relu
    a3 = tf.nn.relu(conv3)
    # max pooling
    p3 = tf.nn.max_pool(a3, ksize=[1,2,2,1], strides=[1,2,2,1], padding="SAME")
    
    #Flatten 
    flat_matrix = tf.reshape(p3, [-1, 15 * 15 * 64])
    
    #FULLY CONNECTED LAYER 1
    l1 = tf.matmul(flat_matrix, lw1) + lb1
    a4 = tf.nn.relu(l1)
    
    # DROP OUT LAYER 1
    keep_prob1 = tf.placeholder(tf.float32)
    layer_drop1 = tf.nn.dropout(a4, keep_prob1)
    
    #FULLY CONNECTED LAYER 2
    l2 = tf.matmul(layer_drop1, lw2) + lb2
    a5 = tf.nn.relu(l2)
    
    # DROP OUT LAYER 2
    keep_prob2 = tf.placeholder(tf.float32)
    layer_drop2 = tf.nn.dropout(a5, keep_prob2)
    
    #FULLY CONNECTED LAYER 3
    l3 = tf.matmul(layer_drop2, wout) + bout
#     a6 = tf.nn.softmax(l3)
    
    
    Z = l3
    
    
    return Z
```

```python
with tf.Session() as sess:
    parameters = initialize_parameters()
    init = tf.global_variables_initializer()
    sess.run(init)
    print(forward_prop(X, parameters))
    sess.close()
```
